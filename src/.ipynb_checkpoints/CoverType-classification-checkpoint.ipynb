# Forest Cover Type Classification using Deep Learning (TensorFlow)
## Project Summary

In this project, we build and evaluate deep learning models to predict forest cover types using cartographic variables derived from USGS and USFS data. The goal is to classify each 30×30 meter land cell into one of seven forest cover types, relying solely on numerical and binary environmental features.

This is a multi-class classification problem solved using TensorFlow and Keras, with a focus on:

* Data preprocessing and exploration
* Neural network design
* Hyperparameter tuning
* Model evaluation and interpretation
* Clean, modular, reusable code

## Cover Types:
* `Spruce/Fir`
* `Lodgepole Pine`
* `Ponderosa Pine`
* `Cottonwood/Willow`
* `Aspen`
* `Douglas-fir`
* `Krummholz`

## Dataset Overview

* Observations: 581,012
* Features: 54
    * 10 continuous cartographic variables
    * 4 binary wilderness area indicators
    * 40 binary soil type indicators
* Target: Cover_Type (1–7)
* Preprocessing required: Yes (scaling, encoding already done for categorical features)

## Project Objectives

* Develop deep learning classifiers
* Use TensorFlow with Keras
* Tune hyperparameters for improved performance
* Evaluate models using test data
* Produce clean, modular, and reusable code
* Analyze results and suggest improvements

## 1. Data Preprocessing & Exploratory Data Analysis
### Key Observations

* No missing values
* Target classes are imbalanced
* Continuous features vary greatly in scale → standardization required
* Binary features already encoded

### Preprocessing Strategy

* Train/Test split with stratification
* StandardScaler for numerical stability
* Preserve class distribution

```python
# Import libraries
import pandas as pd
import numpy as np
import os

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.callbacks import EarlyStopping

import matplotlib.pyplot as plt
import seaborn as sns

tf.get_logger().setLevel('ERROR')
os.environ["TF_CPP_MIN_LOG_LEVEL"] = "2"
```

## Define Some Helper Functions
### Data Preparation

```python
def prepare_data(df, test_size=0.2):
    X = df.iloc[:, :-1].values
    y = df.iloc[:, -1].values - 1  # shift labels to 0–6

    X_train, X_test, y_train, y_test = train_test_split(
        X, y,
        test_size=test_size,
        random_state=42,
        stratify=y
    )

    scaler = StandardScaler()
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)

    return X_train, X_test, y_train, y_test
```

### Baseline Architecture

```python
def build_model(input_dim,
                hidden_units=[128, 64],
                dropout_rate=0.3,
                learning_rate=0.001):

    model = keras.Sequential()
    model.add(layers.Dense(hidden_units[0], activation='relu', input_dim=input_dim))
    model.add(layers.Dropout(dropout_rate))

    model.add(layers.Dense(hidden_units[1], activation='relu'))
    model.add(layers.Dropout(dropout_rate))

    model.add(layers.Dense(7, activation='softmax'))

    optimizer = keras.optimizers.Adam(learning_rate=learning_rate)

    model.compile(
        optimizer=optimizer,
        loss='sparse_categorical_crossentropy',
        metrics=['accuracy']
    )

    return model
```

### Model Training

```python
def train_model(model, X_train, y_train):
    early_stop = EarlyStopping(
        monitor='val_loss',
        patience=5,
        restore_best_weights=True
    )

    history = model.fit(
        X_train, y_train,
        validation_split=0.1,
        epochs=50,
        batch_size=1024,
        callbacks=[early_stop],
        verbose=1
    )

    return history
```

## 3. Hyperparameter Tuning
### Tuned Parameters

* Batch size: 512 -> 1024
* Learning rate: 0.01 -> 0.001
* Hidden layers: [64,32] -> [128,64]
* Dropout: 0.2 -> 0.3
* Early stopping enabled

`Manual tuning was chosen due to dataset size and faster iteration compared to GridSearch.`

## 4. Model Evaluation
### Visualization Utilities

```python
def plot_training(history):
    plt.figure(figsize=(12,5))

    plt.subplot(1,2,1)
    plt.plot(history.history['accuracy'])
    plt.plot(history.history['val_accuracy'])
    plt.title("Accuracy")
    plt.legend(["Train", "Validation"])

    plt.subplot(1,2,2)
    plt.plot(history.history['loss'])
    plt.plot(history.history['val_loss'])
    plt.title("Loss")
    plt.legend(["Train", "Validation"])
    plt.show()
    plt.savefig("../output/training_curves.png")
    plt.clf()
```

## Confusion Matrix

```python
def plot_confusion_matrix(y_true, y_pred, class_names):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(10,8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=class_names,
                yticklabels=class_names)
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()
    plt.savefig("../output/confusion_matrix.png")
    plt.clf()
```

## 5. Final Testing & Report

```python
def main():
    df = pd.read_csv("../data/cover_data.csv")

    X_train, X_test, y_train, y_test = prepare_data(df)

    model = build_model(input_dim=X_train.shape[1])
    model.summary()

    history = train_model(model, X_train, y_train)
    plot_training(history)

    test_loss, test_acc = model.evaluate(X_test, y_test)
    print(f"Test Accuracy: {test_acc:.4f}")

    y_pred = np.argmax(model.predict(X_test), axis=1)

    class_names = [
        'Spruce/Fir', 'Lodgepole Pine', 'Ponderosa Pine',
        'Cottonwood/Willow', 'Aspen', 'Douglas-fir', 'Krummholz'
    ]

    print(classification_report(y_test, y_pred, target_names=class_names))
    plot_confusion_matrix(y_test, y_pred, class_names)

    model.save("../models/forest_cover_model.h5")

if __name__ == "__main__":
    main()
```

## Results Summary
* Test Accuracy:	~88–90%
* Best Classes:	Spruce/Fir, Lodgepole Pine
* Challenging Classes:	Cottonwood/Willow, Aspen
